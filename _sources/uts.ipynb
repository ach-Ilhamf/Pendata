## Menentukan variabel independen dan dependen.
Kolom target dihapus dari data karena akan menjadi variabel dependen.

```python
# Variabel independen
x = data.drop(['target'], axis = 1)
x.head()
```

```python
Y = data['target'].values
Y[0:61]
```

```python
# Variabel dependen
y = data['target']
y.head()
```

## Membagi dataset menjadi training set dan test set.
Dataset dibagi menjadi 80% training set dan 20% test set.

```python
# Import train_test_split function
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2, random_state = 123)
```

## Menentukan prediksi pada training set dan test set.

```python
# Import Gaussian Naive Bayes model
from sklearn.naive_bayes import GaussianNB
# Mengaktifkan/memanggil/membuat fungsi klasifikasi Naive bayes
modelnb = GaussianNB()
# Memasukkan data training pada fungsi klasifikasi naive bayes
nbtrain = modelnb.fit(x_train, y_train)
nbtrain.class_count_
```

#Hasil prediksi dari x_test

```python
# Menentukan hasil prediksi dari x_test
y_pred = nbtrain.predict(x_test)
y_pred
```

## Menentukan nilai probabilitas x_test

```python
# Menentukan probabilitas hasil prediksi
nbtrain.predict_proba(x_test)
```

#Confussion Matrix
Prediksi target dinyatakan gagal dan ternyata target gagal sebanyak 27, prediksi target gagal dan ternyata target berhasil 7, prediksi traetment target dan ternyata target gagal sebanyak 3 dan prediksi target berhasil dan ternyata target berhasil sebanyak 24.

```python
# import confusion_matrix model
from sklearn.metrics import confusion_matrix
confusion_matrix(y_test, y_pred)
```

#Perhitungan nilai akurasi

```python
# Menghitung nilai akurasi dari klasifikasi naive bayes 
from sklearn.metrics import classification_report
print(classification_report(y_test,y_pred))
```

# 2. Metode KNN

```python
import pandas as pd
data = pd.read_csv('https://raw.githubusercontent.com/kb22/Heart-Disease-Prediction/master/dataset.csv')
data
```

##Menampilkan data

```python
data.head()
```

##Mengecek ukuran data

```python
data.shape
```

##Menjadikan dataset sebagai input (X) dan target (Y)

```python
#Membuat dataframe dengan semua training data kecuali kolom target
X = data.drop(columns=["target"])
#mengecek apakah variabel target sudah dihapus
X.head()
```

```python
Y = data["target"].values
Y[0:150]
```

##Membagi dataset menjadi training set dan test set.

```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=1, stratify=Y)
```

##Scaling data

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
scaler.fit(X_train)

X_train_scaled = scaler.transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

##Membuat testing model

```python
from sklearn.neighbors import KNeighborsClassifier
# Membuat KNN classifier
knn = KNeighborsClassifier(n_neighbors = 3)
knn.fit(X_train_scaled,Y_train)
```

##Menampilkan 5 model prediksi pertama pada data test

```python
knn.predict(X_test_scaled)[0:5]
```

##Akurasi model dari data test

```python
knn.score(X_test_scaled, Y_test)
```
